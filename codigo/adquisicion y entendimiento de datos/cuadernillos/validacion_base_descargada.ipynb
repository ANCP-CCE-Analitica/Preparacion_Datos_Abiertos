```python
secop_2023 = secop_2022_2023[(secop_2022_2023['Fecha de Firma']>= '2023-01-01')]
secop_2023['Fecha de Firma'] = pd.to_datetime(secop_2023['Fecha de Firma']).dt.tz_localize(None)
secop_2023 = secop_2023[(secop_2023['Estado']!='Cancelado')]
secop_ii_2023 = secop_2023[(secop_2023['Fuente']=='SECOP II')]
secop_2023_c1 = secop_ii_2023.groupby(['ID Contrato', 'Fecha de Firma'])['Valor del contrato'].max().reset_index()
secop_2023_c1['mes'] = secop_2023_c1['Fecha de Firma'].dt.month
```

```python
secop_2023_c1.groupby(['mes']).agg({'ID Contrato': ['count'], 'Valor del contrato': ['sum']}).reset_index()
```

```python
secop_2023_c2 = secop_ii_2023.groupby(['ID Contrato', 'Modalidad'])['Valor del contrato'].max().reset_index()
```
