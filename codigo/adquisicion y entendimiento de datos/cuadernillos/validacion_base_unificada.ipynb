```python
### Paquetes usados para la validación de datos

import re
import pandas as pd
import numpy as np
from datetime import (
    datetime,
    timedelta,
)
import matplotlib.pyplot as plt
import datetime as dt
import unidecode
import libreria.data_management as dm
from dataprep.clean import clean_duplication
import warnings
warnings.filterwarnings('ignore')
pd.options.display.float_format = '{:.0f}'.format
```

```python
secop_2023 = pd.read_parquet('../../../muestras de datos/procesados/SECOP_2023_depurado.parquet')
secop_2023.head()
```

```python
len(secop_2023)
```

```python
secop_2023['Fecha_firma'] = pd.to_datetime(secop_2023['Fecha_firma']).dt.tz_localize(None)
secop_2023 = secop_2023[(secop_2023['Estado']!='cancelado')]
secop_ii_2023 = secop_2023[(secop_2023['Fuente']=='secop ii')]
secop_2023_c1 = secop_ii_2023.groupby(['ID_Contrato', 'Fecha_firma'])['Valor_contrato'].max().reset_index()
secop_2023_c1['mes'] = secop_2023_c1['Fecha_firma'].dt.month
```

```python
## CANTIDAD DE CONTRATOS Y VALOR POR MES DEL AÑO 2023 ----
## El calculo se realiza para optener los contratos unicos tanto en valor como en cantidad.       
## Los datos corresponden con la información reportada por las entidades estatales en el SECOP II 
secop_2023_c1.groupby(['mes']).agg({'ID_Contrato': ['count'], 'Valor_contrato': ['sum']}).reset_index()
```

```python
secop_2023_c2 = secop_ii_2023.groupby(['ID_Contrato', 'Modalidad'])['Valor_contrato'].max().reset_index()
```

```python
# CANTIDAD DE CONTRATOS Y VALOR POR MODALIDAD DE LOS MESES DEL 1 AL 6 DEL AÑO 2023
secop_2023_c2.groupby(['Modalidad']).agg({'ID_Contrato': ['count'], 'Valor_contrato': ['sum']}).reset_index()
```

```python
secop_2023_c3 = secop_ii_2023.groupby(['ID_Contrato', 'Tipo_de_contrato'])['Valor_contrato'].max().reset_index()
```

```python
## CANTIDAD DE CONTRATOS Y VALOR POR TIPO DE CONTRATO DE LOS MESES DEL 1 AL 6 DEL AÑO 2023
secop_2023_c3.groupby(['Tipo_de_contrato']).agg({'ID_Contrato': ['count'], 'Valor_contrato': ['sum']}).reset_index()
```

```python
secop_2023_c4 = secop_ii_2023.groupby(['ID_Contrato', 'Nombre_Entidad'])['Valor_contrato'].max().reset_index()
```

```python
## TOP 20 - CANTIDAD DE CONTRATOS Y VALOR POR ENTIDAD DE LOS MESES DEL 1 AL 6 DEL AÑO 2023
secop_2023_c4.groupby(['Nombre_Entidad']).agg({'ID_Contrato': ['count'], 'Valor_contrato': ['sum']}).reset_index().sort_values(by=('Valor_contrato',   'sum'),ascending=False).head(20)
```
